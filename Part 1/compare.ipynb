```python
from pathlib import Path

CHAR_PATH = Path("CharacterLevelLM/part1.txt")
BPE_PATH  = Path("TokenizedLM/part1.txt")

def load_labels(p: Path):
    labels = []
    with p.open("r", encoding="utf-8") as f:
        for line in f:
            s = line.strip()
            lab = s[:1].upper() if s else None
            labels.append(lab if lab in ("A","B") else None)
    return labels

char = load_labels(CHAR_PATH)
bpe  = load_labels(BPE_PATH)

N = min(len(char), len(bpe))
valid = diff = 0
for i in range(N):
    if char[i] is None or bpe[i] is None:
        continue
    valid += 1
    if char[i] != bpe[i]:
        diff += 1

pct = 100.0 * diff / valid if valid else 0.0
print(f"Compared lines: {N} (valid A/B pairs: {valid})")
print(f"Differing labels: {diff}  ({pct:.2f}%)")
if len(char) != len(bpe):
    print(f"[note] Different lengths: char={len(char)} vs bpe={len(bpe)}; compared up to {N}.")
```
